```python
#OpenCv is the computer vision library we use for performing all operations
import cv2
import numpy as np

# Here I am loading the images from my local device
square_image = cv2.imread('C:/Users/sandh/OneDrive - University of Florida/Desktop/CV/P_Assignment1/img1.png')
face_image = cv2.imread('C:/Users/sandh/OneDrive - University of Florida/Desktop/CV/P_Assignment1/img2.jpg')

# Get the dimensions of the input images
square_height, square_width, _ = square_image.shape
face_height, face_width, _ = face_image.shape

# Define the affine transformation matrices (replace with your actual matrices)
# I got all these values by decomposing the given matrix. We need to convert all 2x2 matrix into 3x3 for applying translation.
#The sheer value here represents sheer with x axis being 1/3 and y being 1 which will pull the image towards the y-axis. 
Shear = np.array([[1, 1/3, 0], [1, 1, 0], [0, 0, 1]], dtype=np.float32)
#Then we calculate the accurate rotation matrix, in my case I took it as theta=0, cause i was only able to resolve the values then.
theta = np.arctan(0)
Rotation = np.array([[np.cos(theta), -np.sin(theta), 0], [np.sin(theta), np.cos(theta), 0], [0, 0, 1]], dtype=np.float32)
#Here my magnification along x axis is 1 and along y become 3. Which is just streaching the image more towards y.
Magnification = np.array([[1, 0, 0], [0, 3, 0], [0, 0, 1]], dtype=np.float32)
# we finally shift the image by 4 in the x direction. But since 4 is a very tiny when compared with the image's size, we may not be able to see the difference.
Translation = np.array([[1, 0, 4], [0, 1, 0], [0, 0, 1]], dtype=np.float32)


# We use wrap Perspective for performing the operations.This already uses binary interpolation, so we needn't use it seperately I included it in the flags part.
# I have also used square_width and height resizing for displaying the full length image after transformations.
# First we start with shear for square, I have done both of them seperately.
sheared_square = cv2.warpPerspective(square_image, Shear, (square_width * 2, square_height * 2), flags=cv2.INTER_LINEAR)
cv2.imshow('Sheared Square', sheared_square)

# Here we do rotation.
rotated_sheared_square = cv2.warpPerspective(sheared_square, Rotation, (square_width * 2, square_height * 2), flags=cv2.INTER_LINEAR)
cv2.imshow('Rotated Sheared Square', rotated_sheared_square)

# Then magnification or scaling is done.
magnified_rotated_sheared_square = cv2.warpPerspective(rotated_sheared_square, Magnification,(square_width * 6, square_height * 6), flags=cv2.INTER_LINEAR)
cv2.imshow('Magnified Rotated Sheared Square', magnified_rotated_sheared_square)

# this does the translation.
translated_magnified_rotated_sheared_square = cv2.warpPerspective(magnified_rotated_sheared_square, Translation,(square_width * 6, square_height * 6) , flags=cv2.INTER_LINEAR)
cv2.imshow('Translated Magnified Rotated Sheared Square', translated_magnified_rotated_sheared_square)

# This is where we perform the final transformation of that into the given affine matrix.
composed_matrix = np.matmul(Shear, np.matmul(Rotation, np.matmul(Magnification, Translation)))

#I have used this part to get the composed matrix, which will show us how its similar to our main affine transform.
print(composed_matrix)

#show the final transformed square
composed_square = cv2.warpPerspective(square_image, composed_matrix, (square_width * 5, square_height * 5), flags=cv2.INTER_LINEAR)
cv2.imshow('Composed Transformation Square', composed_square)

# Do the same for face
sheared_face = cv2.warpPerspective(face_image, Shear, (face_width * 2, face_height * 2), flags=cv2.INTER_LINEAR)
cv2.imshow('Sheared Face', sheared_face)
#rotation of face image
rotated_sheared_face = cv2.warpPerspective(sheared_face, Rotation, (face_width * 2, face_height * 2), flags=cv2.INTER_LINEAR)
cv2.imshow('Rotated Sheared Face', rotated_sheared_face)
#get the maginified image for face
magnified_rotated_sheared_face = cv2.warpPerspective(rotated_sheared_face, Magnification, (face_width * 4, face_height * 4), flags=cv2.INTER_LINEAR)
cv2.imshow('Magnified Rotated Sheared Face', magnified_rotated_sheared_face)
#do the translation of face image
translated_magnified_rotated_sheared_face = cv2.warpPerspective(magnified_rotated_sheared_face, Translation,(face_width * 4, face_height * 4), flags=cv2.INTER_LINEAR)
cv2.imshow('Translated Magnified Rotated Sheared Face', translated_magnified_rotated_sheared_face)
#show the result of face after all transformation, here the composed_matrix is taken from above.
composed_face = cv2.warpPerspective(face_image, composed_matrix, (face_width * 5, face_height * 5), flags=cv2.INTER_LINEAR)
cv2.imshow('Composed Transformation Face', composed_face)

# Show the original images
cv2.imshow('Square', square_image)
cv2.imshow('Face', face_image)

# Wait for a key press and then close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# We first get all the required libraraies like opencv, numpy and matplot for the display.
import cv2
import numpy as np
import matplotlib.pyplot as plt

# I created a function for the first fourier transform and dividing the results into their magnitude and phase components.
def compute_fourier_magnitude_phase(image):
    # fft2 gives us the fourier transform of the image
    f_transform = np.fft.fft2(image)
    
    # We take out mangintude using abs, which takes out the absolute value from the data. Since magnitude represents strength.
    magnitude = np.abs(f_transform)
    #Then we calculate phase using angle, which gives us the edges and the flow of our image.
    phase = np.angle(f_transform)
    
    return magnitude, phase

#Now we perform inverse fourier. 
def inverse_fourier_transform(magnitude, phase):
    # Combine magnitude and phase to create a complex-valued spectrum
    spectrum = magnitude * np.exp(1j * phase)
    
    # Perform inverse Fourier transform - ifft is inverse fourier.
    reconstructed_image = np.fft.ifft2(spectrum)
    
    # Ensure the real part of the result
    reconstructed_image = np.real(reconstructed_image)
    
    return reconstructed_image

# Get your and your friends image
image_A = cv2.imread("C:/Users/sandh/OneDrive - University of Florida/Desktop/CV/P_Assignment1/img11 (1).jpeg", cv2.IMREAD_GRAYSCALE)
image_B = cv2.imread("C:/Users/sandh/OneDrive - University of Florida/Desktop/CV/P_Assignment1/img11 (2).jpeg", cv2.IMREAD_GRAYSCALE)

# We use the functions to first compute fourier transform to get the mangitude values
magnitude_A, phase_A = compute_fourier_magnitude_phase(image_A)
#here we perform the inverse, but by zeroing the phase and unity magnitude to get the magnitude and phase reconstruted images as per the question.
ma=inverse_fourier_transform(magnitude_A, 0)
pa=inverse_fourier_transform(1, phase_A)
# do the same for 2nd picture.
magnitude_B, phase_B = compute_fourier_magnitude_phase(image_B)
mb=inverse_fourier_transform(magnitude_B, 0)
pb=inverse_fourier_transform(1, phase_B)

#Here we scale or normalize the image to get a better picture as to what happend. Else we will get blank images.
ma_scaled = cv2.normalize(ma, None, 0, 255, cv2.NORM_MINMAX)
mb_scaled = cv2.normalize(mb, None, 0, 255, cv2.NORM_MINMAX)
# Display the original images. I created plot to display the 3X2 pictures.
plt.subplot(2, 3, 1)
plt.imshow(image_A, cmap='gray')
plt.title('Image A')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(image_B, cmap='gray')
plt.title('Image B')
plt.axis('off')

# Display magnitude-only reconstructions
plt.subplot(2, 3, 2)
plt.imshow(np.log1p(ma_scaled), cmap='gray')
plt.title('Magnitude of A')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(np.log1p(mb_scaled), cmap='gray')
plt.title('Magnitude of B')
plt.axis('off')

# Display phase-only reconstructions
plt.subplot(2, 3, 3)
plt.imshow(np.angle(pa), cmap='gray')
plt.title('Phase of A')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(np.angle(pb), cmap='gray')
plt.title('Phase of B')
plt.axis('off')

# This is how i calculate the composite, this is done by interchanging the phase for both the given images.
composite_A = inverse_fourier_transform(magnitude_A, phase_B)
composite_B = inverse_fourier_transform(magnitude_B, phase_A)

# Display the composite images
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.imshow(image_A, cmap='gray')
plt.title('Original Image A')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(composite_A, cmap='gray')
plt.title('Reconstructed Image from (ii)(a)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(image_B, cmap='gray')
plt.title('Original Image B')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(composite_B, cmap='gray')
plt.title('Reconstructed Image from (ii)(b)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize

# Define the pixel coordinates of the four points
point1 = np.array([5, 100, 1])
point2 = np.array([400, 300, 1])
point3 = np.array([500, 490, 1])
point4 = np.array([20, 20, 5])

# Define a function that calculates the error between the proportional components
def objective_function(params):
    Xc, Yc, Zc, Wc = params
    error1 = (Xc - point1[0] * Wc) / (Xc - point4[0] * Wc) - (Xc - point2[0] * Wc) / (Xc - point3[0] * Wc)
    error2 = (Yc - point1[1] * Wc) / (Yc - point4[1] * Wc) - (Yc - point2[1] * Wc) / (Yc - point3[1] * Wc)
    error3 = (Zc - point1[2] * Wc) / (Zc - point4[2] * Wc) - (Zc - point2[2] * Wc) / (Zc - point3[2] * Wc)
    return error1**2 + error2**2 + error3**2

# Initial guess for camera center coordinates
initial_guess = np.array([0, 0, 0, 1])

# Solve for camera center using numerical optimization
result = minimize(objective_function, initial_guess, method='Nelder-Mead')

# Extract the camera center coordinates in homogeneous form
Xc_homogeneous, Yc_homogeneous, Zc_homogeneous, Wc_homogeneous = result.x

# Display the camera center coordinates
print("Camera Center (Homogeneous Coordinates):")
print(f"Xc: {Xc_homogeneous}")
print(f"Yc: {Yc_homogeneous}")
print(f"Zc: {Zc_homogeneous}")
print(f"Wc: {Wc_homogeneous}")
```
